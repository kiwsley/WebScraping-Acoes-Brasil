```python
#importando as bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

```

```python
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service= servico)
```

```python
#acessando o site fundamentus
url = "https://www.fundamentus.com.br/resultado.php"
navegador.get(url)
```

```python
#Mapear processo de coleta de dados
local_tabela = "/html/body/div[1]/div[2]/table"
elemento = navegador.find_element("xpath",local_tabela)
html_tabela = elemento.get_attribute("outerHTML")

#Utilizando o Pandas para realizar a leitura da tabela ajustando os .(ponto) e ,(Virgulas)
tabela = pd.read_html(str(html_tabela), thousands = '.', decimal = ',')[0]
tabela
```

```python
#Realizar o tratamento das informações mais relavantes definidas

#construindo a tabela
#Definindo a coluna papel como index
tabela = tabela.set_index("Papel")

#Filtrando as colunas que serão uilizadas para o modelo

tabela = tabela[['P/L',"P/VP","Div.Yield", 'EV/EBIT', 'ROIC', 'Liq.2meses','Dív.Brut/ Patrim.','Cresc. Rec.5a']]
tabela
```

```python
#filtrando as variáveis de acordo com readme

tabela=tabela[tabela['P/L']>5]
tabela=tabela[tabela['P/L']<20]
tabela
```

```python
#filtrando o P/VP

tabela=tabela[tabela['P/VP']<3]
tabela
```

```python
tabela=tabela[tabela['EV/EBIT']<10]
tabela
```

```python
#A coluna Roic precisou de tratamento, pois estava como objeto

tabela["ROIC"]= tabela['ROIC'].str.replace("%","")
tabela["ROIC"]= tabela['ROIC'].str.replace(".","")
tabela["ROIC"]= tabela['ROIC'].str.replace(",",".")
tabela["ROIC"]= tabela['ROIC'].astype(float)
tabela
```

```python
selic = 10.5
tabela = tabela[tabela["ROIC"]>selic]
tabela

```

```python
tabela=tabela[tabela['Dív.Brut/ Patrim.']<3]
tabela
```

```python
tabela.dtypes
```

```python
tabela["Cresc. Rec.5a"]= tabela['Cresc. Rec.5a'].str.replace("%","")
tabela["Cresc. Rec.5a"]= tabela['Cresc. Rec.5a'].str.replace(".","")
tabela["Cresc. Rec.5a"]= tabela['Cresc. Rec.5a'].str.replace(",",".")
tabela["Cresc. Rec.5a"]= tabela['Cresc. Rec.5a'].astype(float)
tabela.dtypes
```

```python
tabela=tabela[tabela['Cresc. Rec.5a']>10]
tabela
```

```python
tabela = tabela[tabela['Liq.2meses']> 1000000]
tabela
```

```python
tabela["Div.Yield"]= tabela['Div.Yield'].str.replace("%","")
tabela["Div.Yield"]= tabela['Div.Yield'].str.replace(".","")
tabela["Div.Yield"]= tabela['Div.Yield'].str.replace(",",".")
tabela["Div.Yield"]= tabela['Div.Yield'].astype(float)
tabela.dtypes
```
